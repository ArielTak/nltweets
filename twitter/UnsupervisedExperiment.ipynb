# nltweets Topic Modeling

## LDA, basic preprocessing

```python
docs = []
with open('munitweet.txt') as f:
    for line in f.readlines():
        line = line[2:-1]
        line = line.replace('@sfmta_muni', '')
        print(line)
        docs.append(line)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
def build_model(docs):
    vectorizer = CountVectorizer()
    X = docs
    X = vectorizer.fit_transform(X)
    # print(vectorizer.get_feature_names())


    lda = LatentDirichletAllocation()
    return lda.fit(X), X
lda, X = build_model(docs)
```

```python
from collections import defaultdict
threshold = 0.3
topics_to_doc_nums = defaultdict(list)
for i, x in enumerate(X):
    topic_pred = lda.transform(x)[0]
    for topic_num in range(0,10):
        if topic_pred[topic_num] > threshold:
            topics_to_doc_nums[topic_num].append(i)
print(topics_to_doc_nums)
```

```python
from collections import Counter
def get_topic_words(doc_nums, docs):
    all_tokens = ''
    for doc_num in doc_nums:
        all_tokens = all_tokens + docs[doc_num]
    word_counts = Counter(all_tokens.split())
    return word_counts
    
```

```python
for topic_num in range(0,10):
    print("#### topic_num={} ####".format(topic_num))
    doc_nums = topics_to_doc_nums[topic_num]
    for doc_num in doc_nums:
        print(docs[doc_num])
```

```python
for topic_num in range(0,10):
    print("#### topic_num={} ####".format(topic_num))
    doc_nums = topics_to_doc_nums[topic_num]
    print(get_topic_words(doc_nums, docs))
```

## Remove stop words

```python
import spacy
```

```python
from spacy.lang.en.stop_words import STOP_WORDS
print(STOP_WORDS)
```

```python
cleaned_docs = []
for doc in docs:
    tokens = doc.split()
    cleaned_tokens = []
    for token in tokens:
        token = token.lower()
        if token in STOP_WORDS:
            continue
        cleaned_tokens.append(token)
    cleaned_docs.append(' '.join(cleaned_tokens))
print(cleaned_docs)
```

```python
lda, X = build_model(cleaned_docs)
```

```python
from collections import defaultdict
threshold = 0.3
topics_to_doc_nums = defaultdict(list)
for i, x in enumerate(X):
    topic_pred = lda.transform(x)[0]
    for topic_num in range(0,10):
        if topic_pred[topic_num] > threshold:
            topics_to_doc_nums[topic_num].append(i)
print(topics_to_doc_nums)
```

```python
for topic_num in range(0,10):
    print("#### topic_num={} ####".format(topic_num))
    doc_nums = topics_to_doc_nums[topic_num]
    print(get_topic_words(doc_nums, cleaned_docs))
```
